```python
from environs import Env

import lightly_studio as ls
from lightly_studio import db_manager
```

```python
# Read environment variables
env = Env()
env.read_env()

# Cleanup an existing database
db_manager.connect(cleanup_existing=True)
```

```python
# Define data paths
annotations_json = env.path("EXAMPLES_COCO_JSON_PATH", "/path/to/your/dataset/annotations.json")
images_path = env.path("EXAMPLES_COCO_IMAGES_PATH", "/path/to/your/dataset")

# Create a DatasetLoader from a path
dataset = ls.Dataset.create()
dataset.add_samples_from_coco(
    annotations_json=annotations_json,
    images_path=images_path,
    annotation_type=ls.AnnotationType.INSTANCE_SEGMENTATION,
)
```

```python
# Start the GUI in the background
ls.start_gui_background()
ls.check_gui_background()
```

```python
# Add metadata while the GUI keeps running
for i, sample in enumerate(dataset):
    sample.metadata["index_in_dataset"] = i
```

```python
# Stop the background GUI server
ls.stop_gui_background()
```
